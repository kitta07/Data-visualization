```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a CSV file into a DataFrame
data = pd.read_csv('covid_variants_dataset.csv')
```

```python
# Filter the dataset to include only rows for the specified country
country_data = data[data['Country'] == target_country]
```

```python
# Specifying the country to visualize
target_country = "India"
```

```python
# Calculate the total number of cases for each variant in the country
total_cases_by_variant = country_data.groupby('Variant')['Cases'].sum()
```

```python
# Bar graph to visualize the data
plt.bar(total_cases_by_variant.index, total_cases_by_variant.values, color='red')

# Adding title and axis labels to the graph
plt.title(f'Total COVID-19 Cases by Variant in {target_country}', fontsize=14)
plt.xlabel('Variant', fontsize=10)
plt.ylabel('Total Number of Cases', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=30)

# Display the graph
plt.tight_layout()  # Ensures labels fit properly
plt.show()
```
