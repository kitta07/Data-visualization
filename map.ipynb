```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv('covid_variants_dataset.csv')
```

```python
df.head(10)
```

```python
fig=px.sunburst(df, path=['Variant','Country',], values='Cases',
                width=1000, height=800,color='Recovered',title="COVID-19 Variants Distribution by Variants and Country",
                color_continuous_scale='icefire')
fig.update_layout(title_x=0.5)
fig.show()            
```
